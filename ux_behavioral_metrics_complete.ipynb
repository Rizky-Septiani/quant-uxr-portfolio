#### 📈 PILLAR B — Net Promoter Score (NPS) Relational Baseline & Persona Diagnostics
* **Population Diagnostics:** Calculating the ecosystem-wide *Relational NPS* alongside its **95% Confidence Interval ($95\%\text{ CI}$)** utilizing bootstrap resampling methods to ensure empirical data integrity.

* **Granular Segmentation:** Dissecting the brand loyalty matrix across distinct customer personas (`user_segment`: *Beginner, Intermediate, Advanced*).

* **Statistical Inference:** Executing a **One-Way ANOVA** to empirically validate whether the variance in NPS scores across these three user segments is statistically significant or merely a result of random fluctuation.

* **Executive Visualization:** Generating corporate-grade data visualizations optimized entirely for executive boardroom and C-Suite presentations.

````python
# ==============================================================================
# CELL 2 — PILLAR B: NET PROMOTER SCORE (NPS) RELATIONAL DIAMETRIC ANALYSIS
# Methodology: Bootstrap Resampling (95% CI) & One-Way ANOVA Inference
# ==============================================================================
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ── 1. Define Core NPS Calculation Function ──────────────────────────────────
def calculate_nps(scores):
    """Computes the standard Net Promoter Score from a series of 0-10 inputs."""
    promoters = np.sum(scores >= 9)
    detractors = np.sum(scores <= 6)
    total = len(scores)
    return ((promoters - detractors) / total) * 100

# Calculate overall population baseline
overall_nps = calculate_nps(df_tanam["nps_score"])

# ── 2. Bootstrap Resampling for 95% Confidence Interval ──────────────────────
np.random.seed(2025)  # Maintain deterministic stability
boot_nps_distribution = []
n_iterations = 1000

for _ in range(n_iterations):
    boot_sample = df_tanam["nps_score"].sample(frac=1.0, replace=True)
    boot_nps_distribution.append(calculate_nps(boot_sample))

ci_lower = np.percentile(boot_nps_distribution, 2.5)
ci_upper = np.percentile(boot_nps_distribution, 97.5)

# ── 3. Granular Segmentation ──────────────────────────────────────────────────
segments = ["Beginner", "Intermediate", "Advanced"]
segment_nps_results = {}
segment_score_arrays = []

for seg in segments:
    # Filter scores by customer persona segment
    scores = df_tanam[df_tanam["user_segment"] == seg]["nps_score"]
    segment_score_arrays.append(scores.values)
    segment_nps_results[seg] = {
        "nps": calculate_nps(scores),
        "sample_size": len(scores)
    }

# ── 4. Statistical Inference: One-Way ANOVA ──────────────────────────────────
# Testing Null Hypothesis: Mean NPS values across all 3 segments are equal
f_stat, p_value = stats.f_oneway(*segment_score_arrays)

# ── 5. Print Executive Diagnostic Report ──────────────────────────────────────
print("=" * 75)
print("             EXECUTIVE NPS REVENUE & LOYALTY DIAGNOSTIC REPORT")
print("=" * 75)
print(f"  • Overall System NPS : {overall_nps:.2f}%")
print(f"  • 95% Confidence Int : [{ci_lower:.2f}%, {ci_upper:.2f}%] via Bootstrap")
print(f"{'─'*75}")
print("  • Granular Customer Persona Breakdown:")
for seg in segments:
    print(f"    - {seg:<12} (n={segment_nps_results[seg]['sample_size']}) : {segment_nps_results[seg]['nps']:.2f}%")
print(f"{'─'*75}")
print("  • Statistical Hypothesis Inference (One-Way ANOVA):")
print(f"    - F-Statistic : {f_stat:.4f}")
print(f"    - P-Value     : {p_value:.4f}")
if p_value < 0.05:
    print("    - Decision    : REJECT Null Hypothesis. Significant variance detected across segments.")
else:
    print("    - Decision    : FAIL TO REJECT Null Hypothesis. Variance across segments is statistically minor.")
print("=" * 75)

# ── 6. Executive Visualization Generation ────────────────────────────────────

sns.set_theme(style="whitegrid")
plt.figure(figsize=(9, 5))

# Generate clean, C-suite optimized barchart using specified brand palettes
# FIX: Menambahkan hue=segments dan legend=False untuk menghilangkan FutureWarning Seaborn
ax = sns.barplot(
    x=segments,
    y=[segment_nps_results[seg]["nps"] for seg in segments],
    hue=segments,
    palette=["#4A90E2", "#50E3C2", "#B8E986"],
    legend=False,
    edgecolor="black"
)

# Visual polishments
plt.title("Net Promoter Score (NPS) Across User Segments", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("User Persona Segment", fontsize=11, fontweight="bold", labelpad=10)
plt.ylabel("Net Promoter Score (%)", fontsize=11, fontweight="bold", labelpad=10)
plt.ylim(-10, 50) # Standard framing buffer for visibility

# Annotate exact value labels over bars for quick executive parsing
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f"{height:.1f}%",
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center',
                xytext=(0, 8),
                textcoords='offset points',
                fontsize=11, fontweight="bold")

plt.tight_layout()
plt.show()

````

#### 📉 PILLAR C — Transactional Experience Breakdown (CSAT & CES) by Acquisition Channel
* **Channel Efficiency Analytics:** Quantifying the transactional satisfaction rate (**CSAT % Top-2 Box**) and the mean cognitive load index (**CES Mean**), segmented by user acquisition funnels (`acquisition_type`: *Organic, Referral, Paid Ad*).

* **Statistical Hypothesis Testing:** Running an **Independent Two-Sample T-Test** to scientifically validate whether the *Referral* channel yields a significantly superior product experience compared to *Paid Ads*.

* **Strategic Interpretation:** Identifying which acquisition channel generates the highest-quality user engagement to optimize future marketing budget allocations.



````python
# ==============================================================================
# CELL 3 — TASK C: TRANSACTIONAL EXPERIENCE & FRICTION METRICS BY ACQUISITION
# Methodology: Top-2 Box CSAT Conversion, Mean CES, & Independent Two-Sample T-Test
# ==============================================================================
import numpy as np
import pandas as pd
from scipy import stats

channels = ["Organic", "Referral", "Paid Ad"]
channel_diagnostics = {}

# ── 1. Compute Sectional Metrics Across Acquisition Vectors ──────────────────
for chan in channels:
    channel_data = df_tanam[df_tanam["acquisition_type"] == chan]
    total_count = len(channel_data)

    # Calculate CSAT % using the standard Top-2 Box Method (Scores 4 & 5)
    top2_box_csat = np.sum(channel_data["csat_score"].isin([4, 5]))
    csat_percentage = (top2_box_csat / total_count) * 100 if total_count > 0 else 0

    # Calculate the exact Arithmetic Mean for Customer Effort Score (CES)
    ces_mean = channel_data["ces_score"].mean()

    channel_diagnostics[chan] = {
        "csat_top2": csat_percentage,
        "ces_mean": ces_mean,
        "n": total_count,
        "raw_csat": channel_data["csat_score"].values # Saved for statistical validation
    }

# ── 2. Hypothesis Testing: Referral vs. Paid Ad (Independent T-Test) ─────────
# Null Hypothesis: Mean CSAT scores between Referral and Paid Ad cohorts are equal.
referral_csat_raw = channel_diagnostics["Referral"]["raw_csat"]
paidad_csat_raw = channel_diagnostics["Paid Ad"]["raw_csat"]

t_stat, p_val_ttest = stats.ttest_ind(referral_csat_raw, paidad_csat_raw, equal_var=False)

# ── 3. Print Executive Acquisition Channel Report ─────────────────────────────
print("=" * 75)
print("          ACQUISITION CHANNEL TRANSACTIONAL & USABILITY REPORT")
print("=" * 75)
print(f"  • Vector Matrix Performance:")
for chan in channels:
    print(f"    - {chan:<10} (n={channel_diagnostics[chan]['n']:<3}) "
          f"│ CSAT (Top-2 Box): {channel_diagnostics[chan]['csat_top2']:.2f}% "
          f"│ CES (Mean Index): {channel_diagnostics[chan]['ces_mean']:.2f}/7")
print(f"{'─'*75}")
print("  • Hypothesis Testing: Referral vs. Paid Ad Cohorts (Two-Sample T-Test):")
print(f"    - T-Statistic : {t_stat:.4f}")
print(f"    - P-Value     : {p_val_ttest:.4f}")
if p_val_ttest < 0.05:
    print("    - Decision    : REJECT Null Hypothesis. Referral channel yields significantly higher satisfaction.")
else:
    print("    - Decision    : FAIL TO REJECT Null Hypothesis. Sentiment variance between channels is non-significant.")
print("=" * 75)
````

#### 🔄 PILLAR D — Metric Triangulation: Attitudinal Sentiment vs. Actual Behavioral Telemetry
* **Correlation Modeling:** Establishing a statistical correlation model between user attitudinal intent (`nps_score`) and their actual, real-world active engagement (`transactions_30d`).

* **Hypothesis Validation:** Testing the core business assumption: *Do users who express high brand loyalty linearly demonstrate higher trading frequencies in the live environment?*

* **Strategic Product Implication:** Analyzing the psychological implications and business risks of a potential *decoupling effect* (or lack of correlation) on critical roadmap and feature development decisions.


````python
# ==============================================================================
# CELL 4 — TASK D: METRIC TRIANGULATION ANALYSIS (ATTITUDE VS. BEHAVIOR)
# Methodology: Pearson Product-Moment Correlation & Grouped Behavioral Profiling
# ==============================================================================
import numpy as np
import pandas as pd
from scipy import stats

# ── 1. Calculate Core Pearson Correlation Coefficient ────────────────────────
r_coef, p_val_corr = stats.pearsonr(df_tanam["nps_score"], df_tanam["transactions_30d"])

# ── 2. Structural Deep-Dive: Group Behavioral Patterns by NPS Categories ─────
# Map numerical scores to official Net Promoter categories
def get_nps_tier(score):
    if score >= 9:  return "Promoter"
    if score >= 7:  return "Passive"
    return "Detractor"

df_tanam["nps_tier"] = df_tanam["nps_score"].apply(get_nps_tier)

# Compute exact mean trading frequency for each loyalty profile
behavioral_profile = df_tanam.groupby("nps_tier")["transactions_30d"].agg(["mean", "count", "std"])

# ── 3. Print Executive Triangulation Report ──────────────────────────────────
print("=" * 75)
print("          METRIC TRIANGULATION: ATTITUDE VS. BEHAVIOR PROFILE")
print("=" * 75)
print(f"  • Statistical Correlation Matrix (NPS Score × 30-Day Transactions):")
print(f"    - Pearson Correlation Coefficient (r) : {r_coef:.4f}")
print(f"    - Significance Level (P-Value)       : {p_val_corr:.4f}")
print(f"{'─'*75}")
print("  • Behavioral Profile Breakdown Across Loyalty Tiers:")
for tier in ["Promoter", "Passive", "Detractor"]:
    row = behavioral_profile.loc[tier]
    print(f"    - {tier:<10} (n={int(row['count']):<3}) │ Mean Transactions (Last 30 Days): {row['mean']:.2f}")
print("=" * 75)
````

````python
# ============================================================
# CELL 7 — Validated SUS Algorithmic Formula
# Mitigating the most common analytical errors in tech
# ============================================================

print("""
╔══════════════════════════════════════════════════════════════════╗
║               VALIDATED SUS SCORING ALGORITHM                    ║
╠══════════════════════════════════════════════════════════════════╣
║                                                                  ║
║  Raw Item Scale: 1 (Strongly Disagree) to 5 (Strongly Agree)     ║
║                                                                  ║
║  STEP 1: Normalize ODD Items (Positive Valence)                  ║
║  ➔ Normalized Score = Raw_Score − 1                             ║
║  ➔ Normalized Range: 0–4                                         ║
║                                                                  ║
║  STEP 2: Normalize EVEN Items (Negative Valence/Reverse-Scored)  ║
║  ➔ Normalized Score = 5 − Raw_Score                             ║
║  ➔ Normalized Range: 0–4                                         ║
║                                                                  ║
║  STEP 3: Aggregate All 10 Normalized Contributions               ║
║  ➔ Summed Range: 0–40                                            ║
║                                                                  ║
║  STEP 4: Scale to Enterprise Value                               ║
║  ➔ Final SUS Score = Aggregate_Sum × 2.5                        ║
║  ➔ Standardized SUS Range: 0–100                                 ║
║                                                                  ║
║  ⚠️  HIGH-RISK INDUSTRY ERRORS:                                   ║
║  ❌ Averaging raw scores and multiplying the mean by 20.        ║
║  ❌ Summing raw scores directly without executing reverse-scoring. ║
║  ❌ Treating SUS as a linear percentage ratio scale (It is an    ║
║     ordinal percentile-derived distribution, not a ratio!).      ║
║                                                                  ║
╚══════════════════════════════════════════════════════════════════╝
""")
````

````python
# ============================================================
# CELL 8 — Automated SUS Scoring Engine (Reusable Utility)
# ============================================================

def calculate_sus(df, prefix="sus"):
    """
    Computes standardized SUS scores from a raw survey DataFrame.

    Parameters:
    -----------
    df     : pd.DataFrame ➔ Dataset containing columns from 'sus1' to 'sus10'
    prefix : str          ➔ Column name prefix identifier (default: "sus")

    Returns:
    --------
    pd.Series ➔ Standardized SUS scores per respondent scaled from 0 to 100.

    Psychometric Protocol:
    ----------------------
    Odd Items  (1, 3, 5, 7, 9)  = Positive Orientation ➔ Normalized via (Score - 1)
    Even Items (2, 4, 6, 8, 10) = Negative Orientation ➔ Reverse-scored via (5 - Score)
    Aggregate Sum escalated by a factor of 2.5 ➔ Enterprise Range: 0–100
    """
    odd_items  = [f"{prefix}{i}" for i in [1, 3, 5, 7, 9]]
    even_items = [f"{prefix}{i}" for i in [2, 4, 6, 8, 10]]

    # Normalize positive valence vectors: Raw Score - 1
    normalized_odd  = df[odd_items].subtract(1)

    # Execute reverse-scoring algorithm for negative valence vectors: 5 - Raw Score
    normalized_even = df[even_items].apply(lambda x: 5 - x)

    # Concatenate vectors, aggregate row-wise sums, and scale to enterprise metrics
    all_contributions = pd.concat([normalized_odd, normalized_even], axis=1)
    sus_scores        = all_contributions.sum(axis=1) * 2.5

    return sus_scores


def categorize_sus(score):
    """
    Classifies standardized SUS scores leveraging the empirical grading scales
    established by Bangor et al. (2009) and Bangor et al. (2008).
    """
    if score >= 84.1:
        return "A", "Excellent",   "Best Imaginable"
    elif score >= 80.8:
        return "B", "Good",        "Excellent"
    elif score >= 71.4:
        return "C", "Good",        "Good"
    elif score >= 51.7:
        return "D", "OK",          "OK"
    elif score >= 25.1:
        return "F", "Poor",        "Poor"
    else:
        return "F", "Awful",       "Worst Imaginable"


# ── Operational Validation: Single-Respondent Verification ──
sample_respondent = {
    "sus1": 4, "sus2": 2, "sus3": 4,
    "sus4": 1, "sus5": 4, "sus6": 2,
    "sus7": 4, "sus8": 1, "sus9": 5,
    "sus10": 1
}
df_sample = pd.DataFrame([sample_respondent])
calculated_score = calculate_sus(df_sample).values[0]
grade, classification, adjective = categorize_sus(calculated_score)

print("=" * 55)
print("     SINGLE-RESPONDENT ALGORITHMIC VALIDATION LOG")
print("=" * 55)
print("\n  Raw Inventory Inputs:")
for key, val in sample_respondent.items():
    valence = "Positive" if int(key[3:]) % 2 == 1 else "Negative"
    print(f"  {key}: {val}  ({valence} Valence)")
print(f"\n  Standardized SUS Score = {calculated_score:.1f}")
print(f"  Empirical Grade         = {grade}")
print(f"  Acceptability Category  = {classification}")
print(f"  Adjective Rating        = {adjective}")
````

````python
# ============================================================
# CELL 14 — UMUX-Lite Scoring Engine & Empirical SUS Conversion
# Leveraging validated regression equations for cross-scale alignment
# ============================================================

def calculate_umux_lite(df,
                        col_item1="umux1",
                        col_item2="umux2"):
    """
    Computes standardized UMUX-Lite scores from a raw dataset (0–100 range).

    Mathematical Protocol:
    1. Reverse item 2 vector: (8 - raw_score)
    2. Aggregate raw components: item1 + item2_reversed
    3. Normalize and scale: ((total - 2) / 12) * 100
    """
    item1_score = df[col_item1]
    item2_rev   = 8 - df[col_item2]   # Executing reverse-scoring transformation
    total       = item1_score + item2_rev
    umux_score  = ((total - 2) / 12) * 100
    return umux_score


def umux_to_sus_equivalent(umux_score):
    """
    Converts standardized UMUX-Lite scores into an SUS-equivalent metric.
    Derived via the empirical regression model established by Lewis et al. (2013):
    SUS_equivalent = 0.666 * UMUX + 23.922
    Statistical Power: r = 0.96 (exceptionally strong predictive validity)
    """
    return 0.666 * umux_score + 23.922


# ── Synthetic Dataset Generation: UMUX-Lite User Responses ──
np.random.seed(88)
n = 250

df_umux = pd.DataFrame({
    "user_id" : [f"TS{i:04d}" for i in range(1, n+1)],
    "segment" : np.random.choice(
                    ["Novice", "Intermediate", "Advanced"],
                    n, p=[0.55, 0.30, 0.15]),

    # Item 1: System Capabilities Vector (7-Point Likert)
    "umux1"   : np.random.choice(
                    range(1, 8), n,
                    p=[0.02, 0.05, 0.10, 0.15, 0.28, 0.25, 0.15]),

    # Item 2: System Difficulty Vector (7-Point Likert — Negative Valence)
    "umux2"   : np.random.choice(
                    range(1, 8), n,
                    p=[0.15, 0.25, 0.28, 0.15, 0.10, 0.05, 0.02])
})

df_umux["umux_score"]     = calculate_umux_lite(df_umux)
df_umux["sus_equivalent"] = umux_to_sus_equivalent(df_umux["umux_score"])

# Comparative Validation against Empirical Baseline SUS
mean_umux     = df_umux["umux_score"].mean()
mean_sus_conv = df_umux["sus_equivalent"].mean()

print("=" * 65)
print("             ENTERPRISE UMUX-LITE METRIC DIAGNOSIS")
print("=" * 65)
print(f"""
  Standardized UMUX-Lite Score (0–100) = {mean_umux:.2f}
  Predicted SUS Equivalent             = {mean_sus_conv:.2f}
  Empirical Baseline SUS (Cell 7)      = {overall_sus:.2f}
  Scale Deviation                      = {abs(mean_sus_conv - overall_sus):.2f} Points

  Methodological Alignment Status:
  Cross-Scale Calibration is {"CONSISTENT ✅" if abs(mean_sus_conv - overall_sus) < 5 else "INCONSISTENT ⚠️ — Investigation required"}
  (A variance under 5 points confirms metric equivalence across instruments)

  Operational Strategic Playbook:
  ➔ Cycle Sprint 1 : Deploy full 10-Item SUS to establish concrete baseline metrics.
  ➔ Cycle Sprint 2+: Deploy 2-Item UMUX-Lite for low-friction longitudinal tracking.
  ➔ Analytical Phase: Convert UMUX scores to SUS equivalents to preserve unified reporting metrics.
""")

print("  Granular Cross-Segment Usability Summary:")
print(df_umux.groupby("segment")[["umux_score", "sus_equivalent"]]
      .mean().round(2))
````

````python
# ============================================================
# CELL 21 — Comprehensive Cronbach's Alpha Diagnostic Engine
# Execution architecture mirrors SPSS psychometric outputs
# ============================================================

def cronbach_alpha_comprehensive(df, item_cols, scale_name="Scale"):
    """
    Computes system-wide Cronbach's Alpha, including:
    - Aggregate Internal Consistency Reliability
    - Corrected Item-Total Correlations (ITC)
    - Alpha if Item Deleted Variance Tracking
    - Automated Optimization Telemetry

    Parameters:
    -----------
    df         : DataFrame
    item_cols  : list of item column strings
    scale_name : string label for structural output tracking

    Returns:
    --------
    dict containing all computed psychometric reliability telemetry vectors
    """

    data = df[item_cols].dropna()
    n_items = len(item_cols)
    n_resp  = len(data)

    # ── Aggregate Reliability Vector Calculation ─────────────
    # Standard Structural Formula: α = (k/(k-1)) × (1 - Σσ²_i / σ²_total)
    var_items = data.var(axis=0, ddof=1)
    var_total = data.sum(axis=1).var(ddof=1)
    alpha     = (n_items / (n_items - 1)) * (
                    1 - var_items.sum() / var_total
                )

    # ── Corrected Item-Total Correlation (ITC Matrix) ────────
    # Computes the isolated Pearson correlation of each item
    # against the cumulative scale score EXCLUDING the target item.
    item_stats = []
    for col in item_cols:
        other_scores = data.drop(columns=col).sum(axis=1)
        r_itc, p_itc = stats.pearsonr(data[col], other_scores)

        # Variance Delta: Alpha calculations assuming target item erasure
        data_without = data.drop(columns=col)
        n_k          = n_items - 1
        var_k        = data_without.var(axis=0, ddof=1)
        var_tot_k    = data_without.sum(axis=1).var(ddof=1)
        alpha_without = (n_k / (n_k - 1)) * (
                            1 - var_k.sum() / var_tot_k
                        )

        item_stats.append({
            "Item"                        : col,
            "Mean"                        : data[col].mean(),
            "SD"                          : data[col].std(),
            "Corrected Item-Total r"      : r_itc,
            "Alpha if Item Deleted"       : alpha_without,
            "Recommendation"              : (
                "⚠️  Consider Dropping"
                if r_itc < 0.30
                else "✅ Maintain Item"
            )
        })

    df_items = pd.DataFrame(item_stats)

    # ── Qualitative Alpha Mapping Tiers ─────────────────────
    def interpret_alpha(a):
        if a >= 0.90: return "Excellent (> .90)"
        elif a >= 0.80: return "Good (.80–.89)"
        elif a >= 0.70: return "Acceptable (.70–.79)"
        elif a >= 0.60: return "Questionable (.60–.69)"
        elif a >= 0.50: return "Poor (.50–.59)"
        else: return "Unacceptable (< .50)"

    scale_telemetry = {
        "scale_name"      : scale_name,
        "n_respondents"   : n_resp,
        "n_items"         : n_items,
        "alpha"           : alpha,
        "interpretation"  : interpret_alpha(alpha),
        "item_statistics" : df_items,
        "var_items"       : var_items,
        "var_total"       : var_total
    }

    # ── Executive Report Output Generation (SPSS Format) ────
    print("=" * 65)
    print(f"     RELIABILITY ANALYSIS — {scale_name.upper()}")
    print("=" * 65)
    print(f"""
  RELIABILITY STATISTICS
  ──────────────────────────────────────────────────────
  Cronbach's Alpha         = {alpha:.3f}
  Interpretation           = {interpret_alpha(alpha)}
  N of Items               = {n_items}
  N of Respondents         = {n_resp}
  ──────────────────────────────────────────────────────
  UX Industry Benchmark Standards:
  α ≥ .90 → Excellent (Validated Formal Scale Baseline)
  α ≥ .80 → Good      (Standard Research Publication Baseline)
  α ≥ .70 → Acceptable (Minimum Target for Operational Practice)
  α < .70 → Item Optimization / Revision Mandated
    """)

    print("  ITEM-TOTAL STATISTICS")
    print("  " + "─" * 62)
    print(f"  {'Item':<10} {'Mean':>6} {'SD':>6} "
          f"{'ITC r':>8} {'α if deleted':>14} {'Status'}")
    print("  " + "─" * 62)
    for _, row in df_items.iterrows():
        print(f"  {row['Item']:<10} "
              f"{row['Mean']:>6.2f} "
              f"{row['SD']:>6.2f} "
              f"{row['Corrected Item-Total r']:>8.3f} "
              f"{row['Alpha if Item Deleted']:>14.3f} "
              f"  {row['Recommendation']}")

    return scale_telemetry
````

````python
# ============================================================
# CELL 25 — Item Elimination Simulation: Baseline vs. Optimized Structure
# Highly critical tactical framework frequently prioritized in enterprise interviews!
# ============================================================

def simulate_item_elimination(df, item_cols, scale_name,
                               threshold_itc=0.30):
    """
    Simulates the structural impact of programmatic item dropping based on an ITC threshold.
    Generates a comparative variance matrix mapping alpha changes pre- and post-optimization.
    """
    data          = df[item_cols].dropna()
    n_items_initial = len(item_cols)

    # Core mathematical block for streamlined alpha computation
    def _compute_isolated_alpha(isolated_data):
        k         = isolated_data.shape[1]
        if k < 2: return None
        var_items = isolated_data.var(axis=0, ddof=1).sum()
        var_total = isolated_data.sum(axis=1).var(ddof=1)
        return (k / (k - 1)) * (1 - var_items / var_total)

    alpha_initial = _compute_isolated_alpha(data)

    # Diagnostics loop: Pinpoint sub-optimal indicators violating thresholds
    items_flagged_for_drop = []

    for col in item_cols:
        other_scores = data.drop(columns=col).sum(axis=1)
        r_itc, _ = stats.pearsonr(data[col], other_scores)
        if r_itc < threshold_itc:
            items_flagged_for_drop.append((col, r_itc))

    # Calculate post-optimization reliability metrics
    if items_flagged_for_drop:
        cols_to_drop = [x[0] for x in items_flagged_for_drop]
        items_retained = [c for c in item_cols if c not in cols_to_drop]
        alpha_optimized = _compute_isolated_alpha(data[items_retained])
    else:
        alpha_optimized = alpha_initial
        items_retained  = item_cols

    # Execute Corporate Reporting Output Generation
    print(f"\n{'='*55}")
    print(f"  ITEM ELIMINATION SIMULATION — {scale_name.upper()}")
    print(f"{'='*55}")
    print(f"  ITC Critical Threshold : r < {threshold_itc}")
    print(f"  Baseline Inventory Length : {n_items_initial} Items")
    print(f"  Baseline Scale Alpha      : {alpha_initial:.3f}")

    if items_flagged_for_drop:
        print(f"\n  Indicators Flagged for PROGRAMMATIC DROP:")
        for item, r in items_flagged_for_drop:
            print(f"  → {item}: Isolated ITC r = {r:.3f}")
        print(f"\n  Optimized Inventory Length: {len(items_retained)} Items")
        if alpha_optimized:
            delta = alpha_optimized - alpha_initial
            print(f"  Optimized Scale Alpha     : {alpha_optimized:.3f} "
                  f"({delta:+.3f})")
            print(f"\n  {'✅ Drop Highly Recommended — Substantive Reliability Lift' if delta > 0.02 else '⚠️  Re-evaluate Strategic Priority — Marginal Variance Gain'}")
        else:
            print(f"  Optimized Scale Alpha     : Insufficient remaining items to compute alpha.")
    else:
        print(f"\n  ✅ Scale Configuration Optimal — No items violate requirements.")
        print(f"  (All items clear the ITC baseline threshold of r ≥ {threshold_itc})")

    return items_retained, alpha_optimized


# ── Execute Simulation Pipeline on Aligned SUS Dataset ──────
optimized_items, optimized_alpha = simulate_item_elimination(
    df_sus_aligned,
    sus_columns,
    "SUS TanamSaham",
    threshold_itc=0.30
)
````

### 🌌 5b. Exploratory Factor Analysis (EFA) & Dimensionality Check

````python
# ============================================================
# CELL 26 — EFA Matrix Feasibility: KMO & Bartlett's Test of Sphericity
# Mirrors the exact validation metrics run prior to extraction in SPSS
# ============================================================

print("""
╔══════════════════════════════════════════════════════════════════╗
║             EFA FEASIBILITY CRITERIA — PRE-EXTRACTION            ║
╠══════════════════════════════════════════════════════════════════╣
║                                                                  ║
║  1. KMO (Kaiser-Meyer-Olkin Measure of Sampling Adequacy)       ║
║     → Determines if variance patterns are sufficient for factoring║
║     ≥ 0.90 = Marvelous  | ≥ 0.80 = Meritorious                 ║
║     ≥ 0.70 = Middling   | ≥ 0.60 = Mediocre                    ║
║     < 0.50 = Unacceptable → HALT: Data unsuitable for EFA.       ║
║                                                                  ║
║  2. Bartlett's Test of Sphericity                               ║
║     → H0: The sample correlation matrix is an identity matrix    ║
║        (Meaning: Variables are completely uncorrelated).         ║
║     → Target: p < .05 (Reject H0 = Statistically significant correlations)║
║     → If p > .05 ➔ HALT: Matrix contains insufficient correlation.║
║                                                                  ║
╚══════════════════════════════════════════════════════════════════╝
""")

# Standardize extraction matrix subsetting
data_efa = df_sus_aligned[sus_columns].dropna()

# ── Execute KMO Measure of Sampling Adequacy ────────────────
kmo_all, kmo_model = calculate_kmo(data_efa)

# ── Execute Bartlett's Mathematical Sphericity Test ──────────
chi_square_val, p_value = calculate_bartlett_sphericity(data_efa)

# Qualitative KMO Mapping Logic
def interpret_kmo(kmo_score):
    if kmo_score >= 0.90: return "Marvelous ✅"
    elif kmo_score >= 0.80: return "Meritorious ✅"
    elif kmo_score >= 0.70: return "Middling ✅"
    elif kmo_score >= 0.60: return "Mediocre ⚠️"
    else: return "Unacceptable ❌"

print("=" * 55)
print("     EFA MATRIX FEASIBILITY ANALYSIS")
print("=" * 55)
print(f"""
  KMO Measure of Sampling Adequacy:
  Global KMO Index = {kmo_model:.3f} → {interpret_kmo(kmo_model)}

  Bartlett's Test of Sphericity:
  Chi-Square (χ²)  = {chi_square_val:.2f}
  p-value          = {p_value:.6f} → {"✅ Statistically Significant — Proceed with EFA"
                                       if p_value < 0.05
                                       else "❌ Insufficient Correlation — HALT EFA PIPELINE"}

  Isolated KMO Indices Per Indicator (< 0.50 = Structural Mismatch):
""")

for item, kmo_val in zip(sus_columns, kmo_all):
    status = "✅" if kmo_val >= 0.60 else "⚠️ Flagged for Optimization"
    print(f"  {item}: {kmo_val:.3f} {status}")
````

````python
# ============================================================
# CELL 4 — Pre-Launch Checklist & Pilot Test
# Objective: Create an automated script to scan survey questions
# for common cognitive and structural biases.
# ============================================================

def audit_survey(question_list):
    """
    Automatically audit survey questions to detect common biases.

    Parameters:
    -----------
    question_list : list of dict
        Each dict: {"id": "Q1", "text": "...", "type": "likert/open/yn"}

    Returns:
    --------
    DataFrame containing the audit results
    """
    audit_results = []

    leading_words  = ["great", "good", "excellent", "easy", "new",
                      "sophisticated", "best", "bagus", "baik", "hebat", "mudah"]
    negative_words = ["not", "bad", "difficult", "slow", "hard",
                      "tidak", "buruk", "sulit", "lambat", "susah"]
    double_words   = [" and ", " or ", " as well as ", " along with ", " dan ", " serta ", " atau "]
    recall_words   = ["first time", "back then", "ago", "yesterday", "last week",
                      "last month", "pertama kali", "dulu", "dahulu", "lalu", "kemarin"]

    for q in question_list:
        text = q["text"].lower()
        flags = []

        # Check for leading questions
        if any(word in text for word in leading_words):
            flags.append("⚠️ Potential Leading Question")

        # Check for double-barreled questions
        if any(word in text for word in double_words):
            words_found = [w for w in double_words if w in text]
            flags.append(f"⚠️ Potential Double-Barreled (Target keywords: {words_found})")

        # Check for recall bias
        if any(word in text for word in recall_words):
            flags.append("⚠️ Potential Recall Bias")

        # Check if the question is overly long/wordy
        if len(text.split()) > 25:
            flags.append(f"⚠️ Question too long ({len(text.split())} words)")

        audit_results.append({
            "ID"            : q["id"],
            "Text"          : q["text"][:50] + "...",
            "Type"          : q.get("type", "unknown"),
            "Flags"         : "\n".join(flags) if flags else "✅ OK",
            "Needs Revision": len(flags) > 0
        })

    return pd.DataFrame(audit_results)


# ── Sample survey containing common structural issues ────────
sample_survey = [
    {"id": "Q1", "type": "likert",
     "text": "How often do you use this application?"},
    {"id": "Q2", "type": "likert",
     "text": "How would you rate the transfer feature?"},
    {"id": "Q3", "type": "likert",
     "text": "How great is this new feature we just released?"},   # Leading
    {"id": "Q4", "type": "likert",
     "text": "Is this application fast and easy to use?"},       # Double-barreled
    {"id": "Q5", "type": "open",
     "text": "Can you share your experience when you used this app for the first time?"}, # Recall
    {"id": "Q6", "type": "sus",
     "text": "[SUS Item 1] I think that I would like to use this system frequently."},
    {"id": "Q7", "type": "nps",
     "text": "How likely are you to recommend this application?"},
    {"id": "Q8", "type": "open",
     "text": "What is one thing you would like to change about this app?"},
]

results = audit_survey(sample_survey)

print("=" * 70)
print("     SURVEY AUDIT RESULTS — Pre-Launch Check")
print("=" * 70)
print()
for _, row in results.iterrows():
    status = "🔴" if row["Needs Revision"] else "🟢"
    print(f"  {status} {row['ID']}: {row['Text']}")
    if row["Needs Revision"]:
        for flag in row["Flags"].split("\n"):
            print(f"      → {flag}")
    print()

n_ok = (~results["Needs Revision"]).sum()
n_revision = results["Needs Revision"].sum()
print(f"  Summary: {n_ok}/{len(results)} questions OK | "
      f"{n_revision} questions need revision")
````

````python
# ============================================================
# CELL 6 — Survey Weighting Analytics
# Scenario: Sample over-represents Advanced users
# Weighting = Post-stratification correction for representativeness
# ============================================================

print("""
╔══════════════════════════════════════════════════════════════════╗
║                   WHY SURVEY WEIGHTING MATTERS                   ║
╠══════════════════════════════════════════════════════════════════╣
║                                                                  ║
║  Real-World Corporate Scenario:                                  ║
║  Advanced Users = 13% of the total product population            ║
║  But in our Survey = 15% of total respondents                    ║
║  ➔ Advanced users are over-represented → positive score bias!    ║
║                                                                  ║
║  The Solution: Survey Weighting                                  ║
║  Every respondent is weighted to match actual population proportions║
║  Advanced: Weight = 13% / 15% = 0.867 (down-weighted)            ║
║  Beginner: Weight = 55% / 50% = 1.100 (up-weighted)              ║
║                                                                  ║
║  Psychological Analogy: Post-Stratification Research Methodology ║
║  Adjusting sample distributions to mirror true population frames. ║
║                                                                  ║
╚══════════════════════════════════════════════════════════════════╝
""")

# ACTUAL Population Proportions (Extracted from backend Product Analytics)
proporsi_populasi = {
    "Beginner"    : 0.55,
    "Intermediate": 0.32,
    "Advanced"    : 0.13
}

# Observed SAMPLE Proportions (From our survey dataset)
proporsi_sampel = df_big["segmen"].value_counts(normalize=True)

print("  Population vs. Sample Distribution Comparison:")
print(f"  {'Segment':<15} {'Population':>12} {'Sample':>10} {'Weight':>8}")
print(f"  {'─'*15} {'─'*12} {'─'*10} {'─'*8}")

bobot_dict = {}
for seg in ["Beginner", "Intermediate", "Advanced"]:
    pop  = proporsi_populasi[seg]
    samp = proporsi_sampel.get(seg, 0)
    bbt  = pop / samp if samp > 0 else 1.0
    bobot_dict[seg] = bbt
    print(f"  {seg:<15} {pop:>11.1%} {samp:>9.1%} {bbt:>8.3f}")

# Append post-stratification weights to the master DataFrame
df_big["bobot"] = df_big["segmen"].map(bobot_dict)

# ── Metric Evaluation: Unweighted vs. Weighted Analysis ──────────────────────
def weighted_mean(values, weights):
    return np.average(values, weights=weights)

metrics = ["satisfaction", "csat_score", "sus_score"]
print(f"\n  Impact of Post-Stratification Weighting on Core Metrics:")
print(f"  {'Metric':<15} {'Unweighted':>12} {'Weighted':>10} {'Delta':>9}")
print(f"  {'─'*15} {'─'*12} {'─'*10} {'─'*9}")

for m in metrics:
    uw = df_big[m].mean()
    w  = weighted_mean(df_big[m], df_big["bobot"])
    print(f"  {m:<15} {uw:>12.3f} {w:>10.3f} {w-uw:>+9.3f}")

print(f"""
  💡 Strategic Executive Insight:
  The weighted metrics are noticeably lower. This happens because the influence of
  highly satisfied Advanced users was systematically brought down to match reality.
  ➔ ACTION: Always present the WEIGHTED mean in official executive CPO dashboards!
""")
````

To transition from descriptive metrics to rigorous statistical inference. This phase executes formal hypothesis testing on the **InvestPintar** dataset to determine if user satisfaction levels vary significantly across different acquisition channels, and to quantify the absolute strength of those associations.

```python
# ======================================================================
# INFERENTIAL STATISTICAL ANALYSIS — Hypothesis Testing Engine
# ======================================================================
import pandas as pd
import numpy as np
import scipy.stats as stats

# [Analytical Step] Initialize cross-tabulation and inferential testing on df_big


````python
# ============================================================
# CELL 7 — Chi-Square Test of Independence for Categorical Data
# Objective: Determine if user acquisition channels influence CSAT categories
# ============================================================

print("""
  Chi-Square Analytics in UX Research:
  ───────────────────────────────────────────────────────────────────────
  In Psychology: Evaluates the association between 2 categorical variables.
  In UX Design : The exact same logic! Real-world applications include:
  • Acquisition Channel × Categorized CSAT Scores
  • Device Type × NPS Cohorts (Promoter / Passive / Detractor)
  • User Segment × Feature Adoption Metrics (Adopted / Not Adopted)
  ───────────────────────────────────────────────────────────────────────
""")

# Categorizing Continuous CSAT Scores (1–5 Scale)
df_big["csat_kat"] = df_big["csat_score"].apply(
    lambda x: "Satisfied" if x >= 4 else "Dissatisfied"
)

# Categorizing Continuous Net Promoter Scores (0–10 Scale)
df_big["nps_kat"] = df_big["nps_score"].apply(
    lambda x: "Promoter" if x >= 9
              else ("Passive" if x >= 7 else "Detractor")
)

# ── Contingency Table Matrix: Acquisition Channel × CSAT ──────────────────────
crosstab_csat = pd.crosstab(
    df_big["channel"],
    df_big["csat_kat"],
    margins=True,
    margins_name="Total"
)

print("  CONTINGENCY TABLE: Acquisition Channel × CSAT Performance")
print("  " + "─" * 60)
print(crosstab_csat.to_string())

# Run Chi-Square Contingency Test (Excluding grand totals margins)
ct_no_margin = pd.crosstab(
    df_big["channel"], df_big["csat_kat"]
)
chi2, p_chi, dof, expected = stats.chi2_contingency(
    ct_no_margin
)

# Calculate Cramer's V (Effect size measure for Chi-Square)
n_chi  = ct_no_margin.sum().sum()
cramers_v = np.sqrt(
    chi2 / (n_chi * (min(ct_no_margin.shape) - 1))
)

def interpret_cramers_v(v):
    if v >= 0.50: return "Strong"
    elif v >= 0.30: return "Moderate"
    elif v >= 0.10: return "Weak"
    else: return "Negligible / Trivial"

print(f"""
  STATISTICAL CHI-SQUARE INFERENCE RESUTS
  ──────────────────────────────────────────────────
  χ²({dof}) = {chi2:.3f}
  p-value    = {p_chi:.4f} → {"Statistically Significant ✅" if p_chi < 0.05 else "Not Statistically Significant ❌"}
  Cramer's V = {cramers_v:.3f} → Effect Size Magnitude: {interpret_cramers_v(cramers_v)}

  Strategic Interpretation:
  There is {"a statistically significant" if p_chi < 0.05 else "no statistically significant"} association
  between a user's acquisition channel and their resulting satisfaction classification.
  {"However, because the effect size is " + interpret_cramers_v(cramers_v).lower() + ", the practical difference across channels remains minor." if p_chi < 0.05 and cramers_v < 0.30 else ""}
""")
````

````python
# ============================================================
# CELL 8 — Feature-Level UX Analysis: Performance Diagnostics
# Objective: Identify high-performing features vs. friction points
# ============================================================

fitur_cols = [f"rating_{f.lower()}"
              for f in fitur_names]

# Calculate mean score per feature
fitur_mean = df_big[fitur_cols].mean().round(3)
fitur_mean.index = fitur_names

# Calculate Top-2-Box CSAT percentage (Proportion of responses >= 4)
fitur_csat = {
    f: (df_big[f"rating_{f.lower()}"] >= 4).mean() * 100
    for f in fitur_names
}

# Industry Benchmark: Target CSAT threshold for a high-performing feature
target_csat = 75.0

print("=" * 65)
print("     FEATURE-LEVEL SATISFACTION ANALYSIS — InvestPintar")
print("=" * 65)
print(f"\n  {'Feature':<14} {'Mean':>8} {'CSAT %':>8} "
      f"{'vs Target':>12} {'Status'}")
print(f"  {'─'*14} {'─'*8} {'─'*8} {'─'*12} {'─'*18}")

for fitur in fitur_names:
    mean  = fitur_mean[fitur]
    csat  = fitur_csat[fitur]
    delta = csat - target_csat
    status = ("🟢 Met Benchmark" if csat >= target_csat
              else "🔴 Needs Optimization")
    print(f"  {fitur:<14} {mean:>8.2f} {csat:>7.1f}% "
          f"{delta:>+11.1f}% {status}")

worst_fitur = min(fitur_csat, key=fitur_csat.get)
best_fitur  = max(fitur_csat, key=fitur_csat.get)

print(f"""
  ── Strategic Insights & Diagnostics ─────────────────────────────────
  Top Performing Feature : {best_fitur} (CSAT {fitur_csat[best_fitur]:.1f}%)
  Primary Friction Point : {worst_fitur} (CSAT {fitur_csat[worst_fitur]:.1f}%)
  Performance Variance   : {fitur_csat[best_fitur] - fitur_csat[worst_fitur]:.1f} percentage points
  ──────────────────────────────────────────────────────────────────────
""")
````
